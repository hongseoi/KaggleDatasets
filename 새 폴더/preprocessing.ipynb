```python
import os
import pandas as pd

# csv 파일 경로 설정
csv_folder = "raw"
csv_files = [file for file in os.listdir(csv_folder) if file.endswith(".csv")]

# excel 파일 경로 설정
excel_folder = "raw"
excel_files = [file for file in os.listdir(excel_folder) if file.endswith(".xlsx") or file.endswith(".xls")]

# csv 파일 합치기
csv_dfs = []
for file in csv_files:
    file_path = os.path.join(csv_folder, file)
    csv_dfs.append(pd.read_csv(file_path, encoding='cp949'))

# excel 파일 합치기
excel_dfs = []
for file in excel_files:
    file_path = os.path.join(excel_folder, file)
    excel_dfs.append(pd.read_excel(file_path))

# csv와 excel 데이터프레임 합치기
df = pd.concat(csv_dfs + excel_dfs)

# 결과 확인
print(df)
```

```python
# '측정일시'와 '이산화질소농도(ppm)' 컬럼만 남기기
df = df[['측정일시', '이산화질소농도(ppm)']]

# 결과 확인
print(df)
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'측정일시': 'year', '이산화질소농도(ppm)': 'no2(ppm)'})
```

```python
df.to_csv('no2_in_seoul.csv', index='False')
```
