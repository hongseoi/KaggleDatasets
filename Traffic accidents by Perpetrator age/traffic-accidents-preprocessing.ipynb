```python
forders = os.listdir('/kaggle/input/traffic-accidents-by-perpetrator-age')
print(forders)

int(forders[1][:4])
```

```python
df_all = pd.DataFrame()

for i in range(0, len(forders)):
    if forders[i].split('.')[1] == 'csv':
        file = '/kaggle/input/traffic-accidents-by-perpetrator-age/' + forders[i]
        df = pd.read_csv(file, encoding='cp949')
        df['year'] = int(forders[i][:4])
        df_all = pd.concat([df_all, df])
```

```python
df_all.info()
```

```python
df_all = df_all.rename(columns={
    '가해자연령층':'age',
    '사고건수':'num_of_accident',
    '사망자수':'deathtoll',
    '중상자수':'num_of_sinj',
    '경상자수':'num_of_minj',
    '부상신고자수':'num_of_reported_inj'})
```

```python
df_all
```

```python
df_all['age'].unique()
```

```python
df_all['age'].replace(to_replace=['20세이하'],value=['under20'], inplace=True)
df_all['age'].replace(to_replace=['21-30세'],value=['21~30'], inplace=True)
df_all['age'].replace(to_replace=['31-40세'],value=['31~40'], inplace=True)
df_all['age'].replace(to_replace=['41-50세'],value=['41~50'], inplace=True)
df_all['age'].replace(to_replace=['51-60세'],value=['51~60'], inplace=True)
df_all['age'].replace(to_replace=['61-64세'],value=['61~64'], inplace=True)
df_all['age'].replace(to_replace=['65세이상'],value=['over65'], inplace=True)
df_all['age'].replace(to_replace=['불명'],value=['unknown'], inplace=True)
df_all['age'].replace(to_replace=['65-70세'],value=['65-70'], inplace=True)
df_all['age'].replace(to_replace=['71세이상'],value=['over71'], inplace=True)


df_all['age'].unique()
```
